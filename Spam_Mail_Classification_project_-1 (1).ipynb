

Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Pre-Processing

```python
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

Label Encoding

```python
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
X = mail_data['Message']

Y = mail_data['Category']
```

```python
print(X)
```

Splitting the data into training data & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature Extraction

```python


feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)



Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

Training the Model

Logistic Regression

```python
Logistic_Regression = LogisticRegression()
```

```python
Logistic_Regression.fit(X_train_features, Y_train)
```

Evaluating the trained model

```python
prediction_on_training_data = Logistic_Regression.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
Logistic_Regression.score(X_train_features, Y_train)
```

Training the model

Naive Bayes


```python
Naive_Bayes=MultinomialNB()
```

```python
Naive_Bayes.fit(X_train_features, Y_train)
```

```python
Naive_Bayes.score(X_train_features, Y_train)
```

Building a Predictive System

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]


input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = Logistic_Regression.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('ham mail')
```
